# Pythonでの繰り返し処理

Pythonで繰り返し処理を行う方法について説明します。Pythonでは主に`for`ループと`while`ループの2つの方法を使用して繰り返し処理を実現します。


## 1. `for`ループ

`for`ループは、コレクション内の要素を反復処理するために使用されます。主にリスト、タプル、文字列などのイテラブルなデータ構造を対象にします。基本構文は以下の通りです：

```python
for 要素 in イテラブルオブジェクト:
    # 処理
```

例えば、リスト内の要素を順番に処理する場合のコードは次のようになります：

```python
my_list = [1, 2, 3, 4, 5]
for num in my_list:
    print(num)
```


## 2. `while`ループ

`while`ループは、特定の条件が満たされている間、繰り返し処理を行うために使用されます。基本的な構文は以下の通りです：

```python
while 条件:
    # 処理
```

例えば、変数`count`が10未満の間、繰り返し処理を行うコードは次のようになります：

```python
count = 0
while count < 10:
    print(count)
    count += 1  # countを1ずつ増やす
```

この場合、`count`が10未満の間、`print(count)`が実行され、`count`が1ずつ増加します。ループの条件が満たされなくなると、ループが終了します。


### `break`と`continue`

Pythonの繰り返し処理には、`break`と`continue`という2つの制御文も利用できます。
- `break`: ループを強制的に終了し、次の処理に移ります。
- `continue`: 現在の繰り返しをスキップし、次の繰り返しを開始します。

これらの制御文を使うことで、特定の条件下でループを終了したり、一部の処理をスキップしたりすることができます。

繰り返し処理はプログラムの中で非常に重要であり、同じコードを何度も実行する必要がある場合に役立ちます。`for`ループと`while`ループを適切に活用することで、さまざまなタイプの繰り返し処理を実現できます。
