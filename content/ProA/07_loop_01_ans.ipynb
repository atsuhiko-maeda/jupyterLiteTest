### Q7.2-2

Q7.2-1を参考に、1から100までの整数の値のうち、「3または5の倍数」・「3の倍数」・「5の倍数」・「3の倍数かつ5の倍数」の個数をそれぞれ異なる変数に保存して表示しましょう。また回答欄以外は変更しないでください。

```python
count = 0 #3または5の倍数の合計個数を保存する変数
count3 = 0 #3の倍数の合計個数を保存する変数
count5 = 0 #5の倍数の合計個数を保存する変数
count15 = 0 #3の倍数かつ5の倍数の合計個数を保存する変数
#回答欄ここから

for i in range(1,100+1):#for文はそれぞれで分けて書いても間違いではないが，1つだけで十分なはず
    #3または5の倍数
    if i % 3 == 0 or i % 5 == 0:
        count += 1

    #3の倍数
    if i % 3 == 0:
        count3 += 1

    #5の倍数
    if i % 5 == 0:
        count5 += 1

    #3の倍数かつ5の倍数
    if i % 3 == 0 and i % 5 == 0:
        count15 += 1

#ここまで
print("1から100までの3または5の倍数の個数は",count,"です")
print("1から100までの3の倍数の個数は",count3,"です")
print("1から100までの5の倍数の個数は",count5,"です")
print("1から100までの3の倍数かつ5の倍数の個数は",count15,"です")
```
